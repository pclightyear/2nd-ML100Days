## 練習時間
試著使用 sklearn datasets 的其他資料集 (boston, ...)，來訓練自己的線性迴歸模型，並加上適當的正則話來觀察訓練情形。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
# prepare boston data set
boston = datasets.load_boston()
X = boston.data
print("Data shape: ", X.shape)

x_train, x_test, y_train, y_test = train_test_split(X, boston.target, test_size=0.2)
```

```python
# Apply Lasso on Boston data set 
alpha = [1, 10, 100, 1000]
train_err = []
test_err = []

for a in alpha:
    lasso = linear_model.Lasso(alpha=a)
    lasso.fit(x_train, y_train)
    train_err.append(lasso.score(x_train, y_train))
    test_err.append(lasso.score(x_test, y_test))
```

```python
x = []
for a in alpha:
    x.append(str(a))

plt.title("Lasso on boston data set")
plt.xlabel("Alpha")
plt.ylabel("Mean squared error")
    
plt.plot(x, train_err, color='blue')
plt.plot(x, test_err, color='red')
plt.show()
```

```python
# Apply Ridge on Boston data set 
alpha = [0, 1, 1e1, 1e2, 1e3, 1e4, 1e5, 1e6, 1e7, 1e8]
train_err = []
test_err = []

for a in alpha:
    ridge = linear_model.Ridge(alpha=a)
    ridge.fit(x_train, y_train)
    train_err.append(ridge.score(x_train, y_train))
    test_err.append(ridge.score(x_test, y_test))
```

```python
x = []
for a in alpha:
    x.append(str(a))

plt.title("Bridge on boston data set")
plt.xlabel("Alpha")
plt.ylabel("Mean squared error")
    
plt.plot(x, train_err, color='blue')
plt.plot(x, test_err, color='red')
plt.show()
```
